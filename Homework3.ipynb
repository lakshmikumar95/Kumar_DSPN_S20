```r
# 1. Quantitative data is recorded as numbers and represents an objective measurement,
# and this can be continuous (can take on any value) or discrete data (cannot be meaningfully
# divided into smaller increments). Since quantitative data are always numeric they can be ordered,
# added together, and observation frequency can be counted. So mean, median and measures of spread
# can be calculated which tells us a lot about the type of distribution quantitative data come
# from (density distributions and normal distributions are appropriate here).
# Whereas qualitative data sorts observations in categories and can be either nominal
# (no order between the categories), binary (two categories), or ordinal (ordered).
# Since qualitative data represent individual categories, calculating descriptive statistics
# is difficult to do. Mean, median, and measures of spread cannot be calculated, but the mode
# can be (frequency distributions are appropriate here).
```

```r
# 2. It means that the data follows a normal distribution (a symmetric, bell shaped frequency distribution)
# where the average falls at 100 and most of the data (95%) falls between 70 and 130, which is twice the
# standard deviation (sd = 15) on either side of the mean. The bell-shape of the histogram means that the
# data clusters most around the mean (m = 100) and tapers off the further it gets from the mean.
```

```r
library(ggplot2)
a = 0
b = 6
n1 = 10
n2 = 1000
n3 = 10000

r1 <- ceiling(runif(n1, min = a, max = b))
r2 <- ceiling(runif(n2, min = a, max = b))
r3 <- ceiling(runif(n3, min = a, max = b))

dat1 = data.frame(dice1_length = r1)
ggplot(dat1, aes(dice1_length)) + geom_histogram(bins = 6)

dat2 = data.frame(dice2_length = r2)
ggplot(dat2, aes(dice2_length)) + geom_histogram(bins = 6)

dat3 = data.frame(dice3_length = r3)
ggplot(dat3, aes(dice3_length)) + geom_histogram(bins = 6)
```

```r
mean((1+2+3+4+5+6)/6) # Expected mean
mean(r1) # Mean of 10 dice rolls
mean(r2) # Mean of 1000 dice rolls
mean(r3) # Mean of 10000 dice rolls
```

```r
library(ggplot2)
a = 0
b = 6
m1 = 10000

p1 <- ceiling(runif(m1, min = a, max = b))
p2 <- ceiling(runif(m1, min = a, max = b))
p3 <- ceiling(runif(m1, min = a, max = b))
p4 <- ceiling(runif(m1, min = a, max = b))
p5 <- ceiling(runif(m1, min = a, max = b))
p6 <- ceiling(runif(m1, min = a, max = b))

data1 = data.frame(dice1 = p1)
ggplot(data1, aes(dice1)) + geom_histogram(binwidth = 1)

data2 = data.frame(dice2 = p1 + p2)
ggplot(data2, aes(dice2)) + geom_histogram(binwidth = 1)

data3 = data.frame(dice3 = p1 + p2 + p3)
ggplot(data3, aes(dice3)) + geom_histogram(binwidth = 1)

data4 = data.frame(dice6 = p1 + p2 + p3 + p4 + p5 + p6)
ggplot(data4, aes(dice6)) + geom_histogram(binwidth = 1)
```

```r
ggplot(data1, aes(sample = dice1)) + geom_point(stat = "qq") + ggtitle("Normal Distribution vs. Sample1")
ggplot(data2, aes(sample = dice2)) + geom_point(stat = "qq") + ggtitle("Normal Distribution vs. Sample2")
ggplot(data3, aes(sample = dice3)) + geom_point(stat = "qq") + ggtitle("Normal Distribution vs. Sample3")
ggplot(data4, aes(sample = dice6)) + geom_point(stat = "qq") + ggtitle("Normal Distribution vs. Sample4")

# As all four samples have the same number of throws, n = 10000, the sample with the most number of dice, 6,
# produces a distribution most similar to a normal distribution. We can see this because the qq plot for that
# sample, compared to the other three samples, most closely resembles a straight line.
```
