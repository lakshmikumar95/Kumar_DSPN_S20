# State and Trait Empathy and Alcohol Use

# Background

The purpose of the study is to investigate the relationships of state and trait empathy and alcohol use in a large sample of young adults using ecological momentary assessment (EMA) methodology.

Alcohol is best described as a social drug because it is most often consumed in social contexts such as bars, parties, and social gatherings (Fairbairn & Sayette, 2014; Hendriks et al., 2018). Therefore, identifying relevant social factors in the etiology of alcohol use problems is particularly important and can better inform prevention strategies against risky drinking behavior (Creswell et al., 2012).

The present study will focus on empathy as it relates to alcohol use because of the critical role it plays in social contexts. Empathy is typically defined as the capacity to understand another’s perspective or mental state (cognitive empathy) and the capacity to respond with an appropriate emotion to another’s emotional state (affective empathy) (Hoffman, 1982; Riggio, Tucker & Coffaro, 1989). Additionally, empathy has been construed as both a between-person (trait level) and within-person (state level) construct (Graaff et al., 2015; Nezlek et al., 2001; Nezlek et al., 2007). Trait empathy can be thought of as a general tendency for a person to show empathy, whereas state empathy can be thought of as a person’s fluctuations in empathy in response to particular interpersonal interactions (Graaff et al., 2015; Nezlek et al., 2001) or features of the social context (Zaki, 2014). Positive affect, affiliation, and social desirability typically motivate people to approach empathy, whereas competition, material costs, and cognitive costs typically motivate people to avoid empathy (Cameron et al., 2019; Zaki, 2014). Thus, though people show individual differences in average empathy (trait empathy), they can also fluctuate and shift meaningfully in their empathy across situations (state empathy). Importantly, state and trait empathy have been shown to function mathematically independent of one other, with each explaining some of the variance in empathy (Nezlek et al., 2007).

Results from both clinical and non-clinical samples suggest that deficits in trait empathy are associated with increased alcohol consumption and more alcohol-related problems (Dethier & Blairy, 2012; Erol et al., 2016; Gizewski et al., 2013; Kornreich et al., 2002; Kornreich et al., 2013; Martinotti et al., 2009; Maurage et al., 2011a; Maurage et al., 2011b; Mohagheghi et al., 2015; Nguyen, Clark, & Belgrave, 2011; Thoma et al., 2013). Additionally one study found that deficits in affective (but not cognitive) empathy were associated with binge drinking (Laghi et al., 2019a! We are unaware of prior studies that examined the relationships between state empathy and alcohol use or problems, however. Research on related constructs highlights why this work might be important. For instance, fluctuating mood states, which are context dependent like empathy, have been shown to robustly influence the direction, intensity, and persistence of a wide range of human behaviors (Holzahauer et al., 2020; Gardner, 1985; Gendolla, 2000; Gendolla & Brinkmann, 2005). Additionally, meaningful within‐person variability in the Big Five personality traits has helped to explain important trait-content manifestations of behavior (Ching et al., 2014; Fleeson, 2007; Fleeson & Glallagher, 2009; Geukes et al., 2017). In general, more research is needed to explore associations between empathy and alcohol use and, in particular, the extent to which state empathy and affective (versus cognitive) empathy is linked to alcohol use. 

The goal of this study is to advance knowledge about the association between empathy and alcohol use by examining these variables in a well-powered (n = 1080) sample of young adult drinkers using ecological momentary assessment (EMA). This study uses daily reports of alcohol use, thereby eliminating recall bias, and empathy engagement, thus providing a measure of state-level empathy. As such, this study provides the first opportunity to disentangle the effects of trait versus state empathy in its associations with alcohol use. Specifically, while prior studies have shown that individuals with low trait empathy (between-person) report more alcohol involvement (Laghi et al., 2019a; Lannoy et al., 2020; Nguyen, Clark, & Belgrave, 2011), the current study will allow us to determine whether empathy is implicated in the within-person process of alcohol use behavior (e.g., do individuals drink more alcohol on days that they report lower empathy?).

# Variables

#### Demographics: Covariates
- **Gender** (gender): Binary variable ("0" indicating Male, "1" indicating Female)
- **Age** (age): Open-ended response (continuous variable) in years

#### Empathy Measures: Predictors
Participants were first asked if they had inteacted with a person ("Yes"/"No"), and if they said yes, they were sent empathy measure to complete. Daily (state-level) empathy was measured throughout the day using two items corresponding to cognitive empathy and one item corresponding to affective empathy. Each item was rated on a slider scale from 0 (“Not at All”) to 100 (“Extremely”).
- **State cognitive empathy 1** (day.cog.em1): “I considered what the person I interacted with was thinking”
- **State cognitive empathy 2** (day.cog.em2): “I considered what the person I interacted with was feeling”
- **State affective empathy** (day.aff.em): “When the person I interacted with showed emotions, I felt their emotions inside of me”

Trait-level empathy was found by averaging cognitive and affective empathy scores across all the days for each participant, resulting in a mean cognitive and affective empathy score for each participant.
- **Trait cognitive empathy 1** (cog.em1): Average of cognitive empathy 1 scores across all days
- **Trait cognitive empathy 2** (cog.em2): Average of cognitive empathy 2 scores across all days
- **Trait affective empathy** (aff.em): Average of affective empathy scores across all days

#### Alcohol Measures: Outcome
Daily alcohol use.
- **Amount drank** (day.alc.amount): How many standard drinks they consumed (open-ended continuous response)

Overall alcohol use.
- **Average alcohol drank** (m.alc.amount): Dividing the total number of standard drinks a participant drank by the number of days they participated in the study.

# Hypothesis

Hypothesis 1: Daily fluctuations in both cognitive and affective state empathy are expected to be associated with fluctuations in same day standard alcoholic drinks consumed.

State cognitive empathy: 

$H_0: \beta = 0$

$H_a: \beta \neq 0$

State affective empathy:

$H_0: \beta = 0$

$H_a: \beta \neq 0$

Hypothesis: Both cognitive and affective trait empathy are expected to be negatively associated with overall average alcohol use.

Trait cognitive empathy: 

$H_0: \beta = 0$

$H_a: \beta < 0$

Trait affective empathy:

$H_0: \beta = 0$

$H_a: \beta < 0$

Hypothesis 3: Associations between empathy and alcohol use are expected to be stronger for affective versus cognitive empathy.

1. Select data relevant for analysis

    - Empathy Measures:

```r
# Data broken by day for each participant. Participants have as many rows of data as days
# the completed for the study.
ldat <- read.csv(file = "LongDrank.csv")
names(ldat)
wdat <- read.csv(file = "WideDrank.csv")
names(wdat)
```

```r
# I'm pick out the variables that I will be using for my analysis.
library(tidyverse)
ldat1 <- select(ldat, id, gender, age, day, 
                day.alc.count,
                day.cog.em1, day.cog.em2, day.aff.em)
head(ldat1)
wdat1 <- select(wdat, id, gender, age, 
                m.alc.amount,
                cog.em1, cog.em2, aff.em)
head(wdat1)
```

```r
# I start by removing missing variables from each dataset.
# Then I run correlations for the three empathy measures to see if the two cognitive empathy
# measures can be combined into one measure, and to check if the cognitive and affective
# empathy measures are different enough from each other.
ldat2 <- ldat1 %>% na.omit()
ldat3 <- select(ldat2, day.cog.em1, day.cog.em2, day.aff.em)
res1 <- cor(ldat3)
round(res1, 2)
wdat2 <- wdat1 %>% na.omit()
wdat3 <- select(wdat2, cog.em1, cog.em2, aff.em)
res2 <- cor(wdat3)
round(res2, 2)
```

```r
# Since the correlation between the two cognitive empathy measures is >0.80 for both state and trait measures,
# I combine them into one cognitive empathy measure for the analysis.
ldat4 <- mutate (ldat2, day.cog.em = (day.cog.em1 + day.cog.em2)/2)
ldat5 <- select(ldat4, id, gender, age, day,
                day.alc.count,
                day.cog.em, day.aff.em)
head(ldat5)
# I log transformed the average alcohol use outcome variable since it was skewed to the right,
# in order to be able to fit a linear regression.
wdat4 <- mutate (wdat2, cog.em = (cog.em1 + cog.em2)/2,
                log.m.alc = log(m.alc.amount))
wdat5 <- select(wdat4, id, gender, age,
                m.alc.amount, log.m.alc,
                cog.em, aff.em)
head(wdat5)
```

```r
#Summary for demographics
smry <- ldat5  %>% 
    summarise(total_n = length(unique(id)),
              age_prop = mean(age),
              age_sd = sd(age),
              female_prop = mean(gender == '0'),
              days = length(day))
smry
#Summary for alcohol use
smry2 <- ldat5  %>%
    summarise(alcohol_total = sum(day.alc.count),
              alcohol_mean = mean(day.alc.count),
              alcohol_sd = sd(day.alc.count),
              alcohol_min = min(day.alc.count),
              alcohol_max = max(day.alc.count))
smry2

#Summary for empathy measures
smry3 <- ldat5  %>%
    summarise(cog.em_mean = mean(day.cog.em),
              cog.em_sd = sd(day.cog.em),
              aff.em_mean = mean(day.aff.em),
              aff.em_sd = sd(day.aff.em))
smry3
```

## Poisson Regression: State-Level

```r
# Fitting a linear model with cognitive and affective empathy as my predictors and including
# gender and age into the model.
glm.fit1 = glm(formula = day.alc.count ~ day.cog.em + day.aff.em + age + gender, data = ldat5)
summary(glm.fit1)
```

```r
# Plotting the residuals
par(mfrow=c(2,2))
plot(glm.fit1)
```

Q-Q Plot: This is used to validate the assumption that $Y$ is normally distributed, which it is not (as can be seen by the histogram above). The line does not generally resembles a straight line.

```r
# Since outcome is non-negative integer count data, I'm fitting the model to a Poisson regression.
glm.fit2 = glm(formula = day.alc.count ~ day.cog.em + day.aff.em + gender + age, 
               data = ldat5, family = poisson)
summary(glm.fit2)
```

```r
library(lme4)
# The data is hierarchical (days nested within people), so I'm accounting for subject
# by fitting a mixed effects Poisson regression. I start with just cognitive and affective
# empathy as my predictors.
glmer.fit1 <- glmer(day.alc.count ~ day.cog.em + day.aff.em + 
                    (1|id), data = ldat5, family = poisson(link = "log"))
summary(glmer.fit1)
```

```r
glmer.fit2 <- glmer(day.alc.count ~ day.cog.em + day.aff.em + gender + age +
                      (1|id), data = ldat5, family = poisson(link = "log"))
summary(glmer.fit2)
```

```r
ic = AIC(glm.fit1, glm.fit2, glmer.fit1, glmer.fit2)
ic
diff(ic$AIC)
```

Looking at the AIC, the difference in AIC very large comparing the models that do not take into account participant ID (accounting for random effect), so it looks like the mixed-effects model accounts for much more variance in number of standard drinks consumed. than the simpler ones. Additionally, within the mixed-effects models, accounting for gender and age accounted for slightly more variance in number of standard drinks consumed.

```r
# Plotting the coefficients of the poisson mixed-effects model.
library(jtools)
library(ggstance)
library(sjPlot)
plot_summs(glmer.fit1, glmer.fit2, scale = TRUE, exp = TRUE)
```

```r
# Plotting number of standard drinks against cognitive empathy and affective empathy.
par(mfrow=c(1,2))
ggplot(ldat5, aes(day.cog.em, day.alc.count, color = gender)) + 
geom_point()
ggplot(ldat5, aes(day.aff.em, day.alc.count, color = gender)) + 
geom_point()
```

## Linear Regression: Trait-Level

```r
lm.fit1 = lm(log.m.alc ~ cog.em + aff.em, dat = wdat5)
summary(lm.fit1)
```

```r
lm.fit2 = lm(log.m.alc ~ cog.em + aff.em + gender + age, dat = wdat5)
summary(lm.fit2)
```

```r
ic = AIC(lm.fit1, lm.fit2)
ic
diff(ic$AIC)
```

Looking at the AIC, the difference in AIC is not that big, but the more complex model accounts for slightly more variance in log of average alcohol use than the simpler model.

```r
# Plotting the residuals
par(mfrow=c(2,2))
plot(lm.fit1)
```

Residuals vs. Fitted values: We are looking for a generally flat line here, which would indicate a  linear data set. Since we have a generally flat line, the linearity assumption is not violated.

Q-Q Plot: This is on the residuals and is used to validate the assumption that $Y$ is normally distributed. The residuals are normalized to be standard normal (i.e., z-score transformed). Since the line generally resembles a straight line, the normality assumption is not violated.

Scale-location: This is similar to the first plot, but the residuals are rescaled to be positive. This helps you to see if any non-linearities in the model are driven by tails in the residual distribution. This plot is how you can check to see if the variance is spread equally along the ranges of input. It’s good if you see a horizontal line with equally (randomly) spread points. Again, we see a generally flat line with randomly spread points, so there does not seem to be non-linearities in the model.

Residuals vs. Leverage: Here we apply the leverage statistic to see if there are any high-leverage values that are driving the model fits. You want to look for clusters of extreme points in the x-dimension here (anomalies in X). Looking at the top left and bottom left of the plot, there does not seem to be any data points that have large leverage on the model fits. We see a generally flat horizontal line, so there does not seem to be any influential (high-leverage) values.

## Bootstrap: State-Level

```r
library(ISLR)
library(boot)
library(bootstrap)
```

```r
mySumm2 <- function(.) {
    c(beta=fixef(.), sig01=sqrt(unlist(VarCorr(.))))
}
set.seed(2)
system.time( boo01 <- bootMer(glmer.fit2, mySumm2, nsim = 250))
```

```r
head(as.data.frame(boo01))
print(boo01)
```

```r
# Intercept
(bCI.1 <- boot::boot.ci(boo01, index=1, type=c("norm", "basic", "perc")))# beta0

# Residual standard deviation:
(bCI.2  <- boot::boot.ci(boo01, index=2, type=c("norm", "basic", "perc")))#beta1
(bCI.3  <- boot::boot.ci(boo01, index=3, type=c("norm", "basic", "perc")))#beta2
(bCI.4  <- boot::boot.ci(boo01, index=4, type=c("norm", "basic", "perc")))#beta3
(bCI.5  <- boot::boot.ci(boo01, index=5, type=c("norm", "basic", "perc")))#beta4
# Among-batch variance:
(bCI.6 <- boot::boot.ci(boo01, index=6, type=c("norm", "basic", "perc"))) # sig01

confint(boo01)
confint(boo01,type="norm")
confint(boo01,type="basic")
```

```r
par(mfrow=c(1,2))
hist(boo01$t[,2], xlab="Cognitive Empathy")
hist(boo01$t[,3], xlab="Affective Empathy")
#we get a distribution of all of the estimates for cognitive and affective empathy
```

## Bootstrap: Trait-Level

```r
# Bootstrap 95% CI for regression coefficients
library(boot)
# function to obtain regression weights
bs <- function(formula, data, indices) {
  d <- data[indices,] # allows boot to select sample
  fit <- lm(formula, data = d)
  return(coef(fit))
}
# bootstrapping with 1000 replications
results <- boot(data = wdat5, statistic = bs,
   R = 1000, formula = log.m.alc ~ cog.em + aff.em + gender + age)

# view results
results
par(mfrow=c(3,3))
plot(results, index=1) # intercept
plot(results, index=2) # cognitive empathy
plot(results, index=3) # affective empathy
plot(results, index=4) # gender
plot(results, index=5) # age
```

```r
# get 95% confidence intervals
boot.ci(results, type="bca", index=1) # intercept
boot.ci(results, type="bca", index=2) # cognitive empathy
boot.ci(results, type="bca", index=3) # affective empathy
boot.ci(results, type="bca", index=4) # gender
boot.ci(results, type="bca", index=5) # age
```

## Bayes Factor: State-Level

```r
library(bayestestR)
m0 <- glmer(day.alc.count ~ (1|id), data = ldat5, family = poisson(link = "log"))
m1 <- glmer(day.alc.count ~ day.cog.em + (1|id), data = ldat5, 
            family = poisson(link = "log"))
m2 <- glmer(day.alc.count ~ day.aff.em + (1|id), data = ldat5, 
            family = poisson(link = "log"))
m3 <- glmer(day.alc.count ~ day.cog.em + day.aff.em + (1|id), data = ldat5, 
            family = poisson(link = "log"))
m4 <- glmer(day.alc.count ~ day.cog.em + day.aff.em + gender + (1|id), 
            data = ldat5, family = poisson(link = "log"))
m5 <- glmer(day.alc.count ~ day.cog.em + day.aff.em + gender + age + (1|id), 
            data = ldat5, family = poisson(link = "log"))

bayesfactor_models(m1, m2, m3, m4, m5, denominator = m0)
```

## Bayes Factor: Trait-Level

```r
library(ISLR)
library(BayesFactor)
full <- lmBF(log.m.alc ~ cog.em + aff.em + gender + age, data = wdat5) #all predictors + interaction
onlyCognitive <- lmBF(log.m.alc ~ cog.em, data = wdat5) #only dose
onlyAffective <- lmBF(log.m.alc ~ aff.em, data = wdat5) #only supplementation
onlyEmpathy <- lmBF(log.m.alc ~ cog.em + aff.em, data = wdat5) #no interaction
allBFs <- c(full, onlyCognitive, onlyAffective, onlyEmpathy)
allBFs
```
