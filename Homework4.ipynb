1a.

$MSE(\beta_0, \beta_1) = E[(Y - (\beta_0 + \beta_1X))^2]$

$MSE(\beta_0, \beta_1) = E[(Y - \beta_0 - \beta_1X)(Y - \beta_0 - \beta_1X)]$

$MSE(\beta_0, \beta_1) = E[(Y^2 - \beta_0Y - \beta_1XY - \beta_0Y + \beta_0^2 + \beta_0\beta_1X - \beta_1XY + \beta_0\beta_1X + \beta_1^2X^2)]$

$MSE(\beta_0, \beta_1) = E[(Y^2 - 2\beta_0Y - 2\beta_1XY + \beta_0^2 + 2\beta_0\beta_1X + \beta_1^2X^2)]$

$MSE(\beta_0, \beta_1) = E[Y^2] - 2\beta_0E[Y] - 2\beta_1E[XY] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2E[X^2])$

$MSE(\beta_0, \beta_1) = E[Y^2] - 2\beta_0E[Y] - 2\beta_1(Cov[XY] + E[X]E[Y]) + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2(Var[X] + (E[X])^2)$

$MSE(\beta_0, \beta_1) = E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[XY] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2$

1b.

$MSE(\beta_0, \beta_1) = E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[XY] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2$

$\frac{\partial E[(Y - (\hat{\beta}_0 + \hat{\beta}_1X))^2]}{\partial \hat{\beta}_0} = - 2E[Y] + 2\beta_0 + 2\beta_1E[X]$

$\frac{\partial E[(Y - (\hat{\beta}_0 + \hat{\beta}_1X))^2]}{\partial \hat{\beta}_0} = 0$

$0 = - 2E[Y] + 2\beta_0 + 2\beta_1E[X]$

$- 2\beta_0 = - 2E[Y] + 2\beta_1E[X]$

$2\beta_0 = 2E[Y] - 2\beta_1E[X]$

$\beta_0 = E[Y] - \beta_1E[X]$

1c.

$MSE(\beta_0, \beta_1) = E[Y^2] - 2\beta_0E[Y] - 2\beta_1Cov[XY] - 2\beta_1E[X]E[Y] + \beta_0^2 + 2\beta_0\beta_1E[X] + \beta_1^2Var[X] + \beta_1^2(E[X])^2$

$\frac{\partial E[(Y - (\hat{\beta}_0 + \hat{\beta}_1X))^2]}{\partial \hat{\beta}_1} = - 2Cov[XY] - 2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

$\frac{\partial E[(Y - (\hat{\beta}_0 + \hat{\beta}_1X))^2]}{\partial \hat{\beta}_1} = 0$

$0 = - 2Cov[XY] - 2E[X]E[Y] + 2\beta_0E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

Plug in $\beta_0 = E[Y] - \beta_1E[X]$

$0 = - 2Cov[XY] - 2E[X]E[Y] + 2(E[Y] - \beta_1E[X])E[X] + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

$0 = - 2Cov[XY] - 2E[X]E[Y] + 2E[X]E[Y] - 2\beta_1(E[X])^2 + 2\beta_1Var[X] + 2\beta_1(E[X])^2$

$0 = - 2Cov[XY] + 2\beta_1Var[X]$

$- 2\beta_1Var[X] = - 2Cov[XY]$

$\beta_1Var[X] = Cov[XY]$

$\beta_1 = \frac{Cov[XY]}{Var[X]}$

2.

$Y = \hat\beta_0 + \hat\beta_1X$

$\beta_0 = E[Y] - \beta_1E[X]$

$\beta_1 = \frac{Cov[XY]}{Var[X]}$

$Y = E[Y] - \beta_1E[X] + (\frac{Cov[XY]}{Var[X]})X$

$Y = E[Y] - (\frac{Cov[XY]}{Var[X]})E[X] + (\frac{Cov[XY]}{Var[X]})X$

$Y = E[Y] + (\frac{Cov[XY]}{Var[X]})(X - E[X])$

Plug $E[X]$ into $X$

$Y = E[Y] + (\frac{Cov[XY]}{Var[X]})(E[X] - E[X])$

$Y = E[Y] + (\frac{Cov[XY]}{Var[X]})(0)$

$Y = E[Y]$

This implies that no matter what the slope and intercept are for the model, the ordinary least squares solution always passes through $(E(X), E(Y))$, which are the response and predictor variable means (suggesting that the response and predictor variable means have a very strong leverage on the OLS solution).

3.

$\hat\beta = \frac{ \sum_{i=1}^{n} (x_i-\bar{x})(y_i - \bar{y})} {\sum_{i=1}^{n} (x_i - \bar{x})^2} = \frac{Cov[XY]}{Var[X]} = \frac{Cov[XY]}{1} = Cov[XY]$

$r = \frac{ \sum_{i=1}^{n} (x_i-\bar{x})(y_i - \bar{y})} {\sum_{i=1}^{n} \sqrt{(x_i - \bar{x})^2} \sqrt{(y_i - \bar{y})^2}} = \frac{Cov[XY]}{\sigma_X\sigma_Y} = \frac{Cov[XY]}{\sqrt{1}\sqrt{1}} = Cov[XY]$

In the case where both $X$ and $Y$ are standard normal distributions (i.e. $X$ and $Y$ have $\mu=0$ and $\sigma=1$) then the OLS solution is the Pearson correlation coefficient. Both are just the covariance of $X$ and $Y$.

4a.

$RSE = \sigma_{model}^2 = \sqrt{ \frac{RSS}{n-2} } = \sqrt{ \frac{ \sum_{i=1}^{n} (y-\hat{y})^2 }{n-2} }$

RSE (Residual Standard Error) is the simplest form of goodness of fit measures as it provides a measure of how much of the observed error $\hat{y}$ explains. The RSE is considered to be a lack of fit of the model to the data, meaning that the smaller the RSE is, the more the model is explaining the variance. The smaller the RSS (Residual Sum of Squares), the smaller the RSE. Ideally, the RSE should be much much smaller than the variance of $Y$ itself ($\sigma_{y}^2$).

4b.

$r^2 = \frac{TSS-RSS}{TSS} = 1 - \frac{RSS}{TSS} = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2} {\sum_{i=1}^{n} (y_i - \bar{y})^2}$

The RSE measures error in the model, whereas the $r^2$ statistic measures the percentage of variance in $Y$ that is explained by $X$. This is the TSS (measure of variance of $Y$) minus RSS (measure of residuals from the model) over TSS. $r^2$ is a ratio, so it is constrained between 0 and 1 (unless it has a very poor model fit), with 0 indicating that $X$ explains 0% of the variance in $Y$, and 1 indiciating that $X$ explains 100% of the variance in $Y$. So the closer the model is to 1, the better it is at explaining the variance in $Y$. The $r^2$ statistic provides a goodness of fit that is meaningful and has interpretable units.

4c.

$F = \frac{ \frac{(TSS-RSS)}{p}} { \frac{RSS}{n - p - 1}}$

The F-statistic is another typical measure of goodness of fit. The F-statistic is a hybrid of the RSE and $r^2$. Unlike $r^2$, which is a ratio that is constrained between 0 and 1, the F-statistic can be any positive number since it is using RSS, rather than TSS, as the denominator. The F-statistic says whether the variance explained by the model (TSS - RSS) is greater than the variance not explained the model (RSS), if so then the ratio is high. We want the ratio to be high.

```r
d <- read.csv(file = "unrestricted_trimmed_1_7_2020_10_50_44.csv")
library(tidyverse)
d1 <- select(d, Subject, Gender, Flanker_Unadj, FS_Total_GM_Vol, FS_Tot_WM_Vol) %>% na.omit()
print(d1)
```

```r
# Load libraries
library(ISLR)
library(ggplot2)

lm.fit1 = lm(Flanker_Unadj ~ FS_Total_GM_Vol, data = d1) # linear model fit
summary(lm.fit1)
coef(lm.fit1) # print regression coefficients
confint(lm.fit1) # print confidence intervals for each regression parameter
```

The coefficients section tells us that total gray matter volume is significantly different from 0 (we can reject the null that $H_0: \beta_1 = 0$). The total gray matter volume effects unadjusted Flanker task scores in a significant manner, and we can see that the direction is positive (each unit change increase in total gray matter volume brings up unadjusted Flanker task scores by 3.110e-05). Additionally, the 95% confidence interval for the FS_Total_GM_Vol does not contain 0, suggesting that we can reject the null and state that there is a significant association between total gray matter volume and unadjusted Flanker task scores.
Our model predicts ~4% of the variance in unadjusted Flanker task scores ($r^2$) and the model predicts more than chance (F-statistic: 49.68, p-value < 0.05).

```r
ggplot(aes(FS_Total_GM_Vol, Flanker_Unadj), data = d1) + 
geom_point() + geom_smooth(method = 'lm')
plot(lm.fit1)
```

Residuals vs. Fitted values: We are looking for a generally flat line here, which would indicate a strictly linear data set. Since we have a generally flat line, the linearity assumption is not violated.

Q-Q Plot: This is on the residuals and is used to validate the assumption that $Y$ is normally distributed. The residuals are normalized to be standard normal (i.e., z-score transformed). Since the line generally resembles a straight line, the normality assumption is not violated.

Scale-location: This is similar to the first plot, but the residuals are rescaled to be positive. This helps you to see if any non-linearities in the model are driven by tails in the residual distribution. This plot is how you can check to see if the variance is spread equally along the ranges of input. Its good if you see a horizontal line with equally (randomly) spread points. Again, we see a generally flat line with randomly spread points, so there does not seem to be non-linearities in the model.

Residuals vs. Leverage: Here we apply the leverage statistic to see if there are any high-leverage values that are driving the model fits. You want to look for clusters of extreme points in the x-dimension here (anomalies in X). Looking at the top left and bottom left of the plot, there does not seem to be any data points that have large leverage on the model fits. We see a generally flat horizontal line, so there does not seem to be any influential (high-leverage) values.

From the first, there does not seem to be any obvious outliers that are not easily described by the best fit line that describe the rest of the data.

Additionally, since there is only one predictor or X variable (FS_Total_GM_Vol), collinearity assumption does not seem to be violated.

```r
GM_values = data.frame(FS_Total_GM_Vol = c(612432, 702562, 596433, 695982, 649230)) 
prediction_table = data.frame(predict(lm.fit1, GM_values, interval = 'confidence'))
predicted_table = cbind(GM_values, prediction_table)
colnames(predicted_table) = c('GM_input', 'predicted fit', 'lower_CI_bound', 'upper_CI_bound') 
round(predicted_table, 2)
```

The predicted Flanker_Unadj values for each of the five FS_Total_GM_Vol values are returned, along with the lower (lwr) and upper (upr) bounds of the 95% confidence interval of those predictions (i.e., we are 95% confident that the real value rests between those two bounds). For example, when FS_Total_GM_Vol = 612432, we are 95% confident that the unadjusted Flanker task score value will be between 108.44 and 110.16.

```r
lm.fit2 = lm(Flanker_Unadj ~ FS_Tot_WM_Vol + FS_Total_GM_Vol + Gender, data = d1)
summary(lm.fit2)
```

```r
lm.fit3 = lm(Flanker_Unadj ~ - FS_Tot_WM_Vol + FS_Total_GM_Vol - Gender, data = d1)
summary(lm.fit3)
```

In the first model and from coefficients table, we can see that FS_Tot_WM_Vol and Gender are non-significant predictors, meaning we do not know if they will impact unadjusted Flanker task scores (but we cannot fully rule out the possibility that they don't). When FS_Tot_WM_Vol and Gender are included, our model predicts 4.14% of the variance in unadjusted Flanker task scores ($r^2$), RSE is 9.866, and the model predicts more than chance (F-statistic: 17.02, p-value < 0.05). Our model fit statistics ($r^2$, F-statistic, RSE) all look to slightly improve when FS_Tot_WM_Vol and Gender are removed. Then our model predicts 4.19% of the variance in unadjusted Flanker task scores ($r^2$), RSE is 9.864, and the model predicts more than chance (F-statistic: 49.68, p-value < 0.05).

```r
lm.fit4 = lm(Flanker_Unadj ~ FS_Total_GM_Vol + Gender + FS_Total_GM_Vol:Gender, data = d1)
summary(lm.fit4)
coef(lm.fit4)
confint(lm.fit4)
```

Adding the interaction term changes the interpretation of all the coefficients. The interaction term between Gender and total gray matter volume is not a significant predictor, suggesting that the effect of total gray matter volume on unadjusted Flanker task scores are not different across genders (but we cannot fully rule out the possibility that they are different). Also, we can see that the 95% confidence interval for this interaction term contains 0. When the FS_Total_GM_Vol and Gender interaction term is included, our model predicts 4.11% of the variance in unadjusted Flanker task scores ($r^2$), RSE is 9.868, and the model predicts more than chance (F-statistic: 16.89, p-value < 0.05). Compared to the above two models, this model shows to have slightly worse model fit statistics.
